## 2.8 그래프 그리기

```python
import numpy as np
from matplotlib import pyplot as plt
```

## 6.1 Plot

```python
epoch = np.arange(1,100)
train_loss = 1.5/(epoch)
val_loss = .3+1/(epoch)
```

```python
plt.figure(figsize=(10,5))
plt.plot(train_loss, 'r-*')
plt.plot(val_loss, 'b-*')
plt.legend(['train','validation'])
plt.title("MNIST")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```

```python
plt.figure(figsize=(5,10))
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'ro', t, t**2, 'bs', t, 5*np.sin(t), 'g^')
plt.legend(['t','t**2','5sin(t)'])
plt.show()
```

## 6.2 Subplot

```python
names = ['First', 'Business', 'Economy']
prices = [20, 5, 1]

plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.bar(names, prices)
plt.title(f"{names[0]}", fontsize=14)

plt.subplot(132)
plt.scatter(names, prices)
plt.xlabel('class', fontsize=12)

plt.subplot(133)
plt.plot(names, prices, linewidth=5.0)
plt.suptitle('Ticket Class', fontsize=14)
plt.axis('off')
plt.show()
```

추가적으로 데이터 분석에서는 seaborn 라이브러리도 많이 사용된다.

```python
random_images = np.random.normal(0, 1, (4, 216, 216))
extend = [-1, 1, -1, 1]
plt.figure(figsize=(5, 5))
plt.subplot(221)
plt.imshow(random_images[0], interpolation='nearest', cmap='jet')
plt.title("t=0", fontsize=20)
plt.clim(-1, 1)
plt.axis('off')

plt.subplot(222)
plt.imshow(random_images[1], interpolation='nearest', cmap='jet')
plt.title("t=1", fontsize=20)
plt.clim(-1, 1)
plt.axis('off')

plt.subplot(223)
plt.imshow(random_images[2], interpolation='nearest', cmap='jet')
plt.title("t=2", fontsize=20)
plt.clim(-1, 1)
plt.axis('off')

plt.subplot(224)
plt.imshow(random_images[3], interpolation='nearest', cmap='jet')
plt.title("t=3", fontsize=20)
plt.clim(-1, 1)
plt.axis('off')

plt.show()
```
